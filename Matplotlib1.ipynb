Ejemplo: Trazar una marca circular en la posición (5,5)

Primero importamos la interfaz de scripting como **plt**

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(5,5,'o') # Hace el trazado
plt.show() # se muestra la grafica
```

La funcion magica %matplotlib inline con el kernel *inline* se utiliza para representar las graficas dentro del navegador. 
Una de sus limitaciones es que una vez trazada la gráfica, esta no se puede modificar. Si se modifica entonces genera un nueva gráfica

```python
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
plt.plot(5,5,'o')
plt.ylabel("Y")
plt.xlabel("X")
plt.title("Trazado de ejemplo")
plt.show()
```

Para superar las limitaciones del backend *inline* es usar el backend *notebook*

```python
%matplotlib notebook
import matplotlib.pyplot as plt
```

```python
plt.plot(5,5,'o')
```

```python
plt.ylabel("Y")
plt.xlabel("X")
plt.title("Ejemplo de trazado")
```
